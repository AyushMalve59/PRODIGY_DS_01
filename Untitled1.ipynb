```python
dataset_final=dataset.drop(['Country Code',	'Indicator Name'],axis=1)
```

```python
dataset.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
chosen_country = 'India'


country_data = dataset_final[dataset_final['Country Name'] == chosen_country]


years = country_data.columns[2:].astype(int)
population = country_data.iloc[:, 2:].values.flatten()

plt.figure(figsize=(5, 3))
plt.barh(years, population, color='pink')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title(f'Population of {chosen_country} Over the Years')
```

```python
plt.hist(population, bins=20, color='skyblue', edgecolor='black')
plt.show()
```
